```python
q1=df['Income'].quantile(0.25)
q3=df['Income'].quantile(0.75)
IQR=q3-q1
```

```python
Income_lower_boundary=q1-1.5*IQR
Income_upper_boundary=q3+1.5*IQR
```

```python
clean_df=df[(df['Income']>Income_lower_boundary) & df['Income']<Income_upper_boundary]
```

```python
clean_df['Income'].plot(kind='box')
```

```python
stats.probplot(clean_df['Income'],dist='norm',plot=plt)
plt.grid()
```
